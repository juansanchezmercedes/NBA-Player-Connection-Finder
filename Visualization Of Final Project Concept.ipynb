```python
import networkx as nx
import matplotlib.pyplot as plt

# Sample data
edges = [
    ('Michael Jordan', 'Juwan Howard', {'weight': 2}),
    ('Juwan Howard', 'LeBron James', {'weight': 1}),
    ('Michael Jordan', 'LeBron James', {'weight': 3})  # Assume 3 shared attributes
]

# Create a directed graph
G = nx.DiGraph()

# Add edges along with weights
G.add_edges_from(edges)

# Position nodes using the spring layout
pos = nx.spring_layout(G)

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_color='lightblue')

# Draw edges with edge labels
nx.draw_networkx_edges(G, pos, width=[G[u][v]['weight'] for u, v, d in edges])
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['weight'] for u, v, d in edges})

# Draw node labels
nx.draw_networkx_labels(G, pos)

# Display
plt.title('Player Connections Visualization')
plt.axis('off')  # Turn off the axis
plt.show()
```
